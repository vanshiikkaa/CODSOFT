```python
# Convert 'trans_date_trans_time' to datetime format
df['new_trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
```

```python
# Extract useful features from 'trans_date_trans_time'
df['trans_year'] = df['new_trans_date_trans_time'].dt.year
df['trans_month'] = df['new_trans_date_trans_time'].dt.month
df['trans_day'] = df['new_trans_date_trans_time'].dt.day
df['trans_hour'] = df['new_trans_date_trans_time'].dt.hour
```

```python
# Convert 'dob' to datetime and calculate the age of the customer
df['new_dob'] = pd.to_datetime(df['dob'])
df['dob_year'] = df['new_dob'].dt.year
df['age'] = df['trans_year'] - df['dob_year']
```

```python
df.head()
```

```python
# Drop original datetime and dob columns 
df = df.drop(columns=['trans_date_trans_time', 'dob'])
```

```python
df.head()
```

```python
#encoding categorical features
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#gender
df['num_gender'] = le.fit_transform(df['gender'])

#category
df['num_cat'] = le.fit_transform(df['category'])

#state
df['num_state'] = le.fit_transform(df['state'])

#merchant
df['num_mer'] = le.fit_transform(df['merchant'])

#city
df['num_city'] = le.fit_transform(df['city'])

#street
df['num_street'] = le.fit_transform(df['street'])

#job
df['num_job'] = le.fit_transform(df['job'])

#trans_num
df['new_trans_num'] = le.fit_transform(df['trans_num'])

#first
df['num_first'] = le.fit_transform(df['first'])

#last
df['num_last'] = le.fit_transform(df['last'])

df.head()
```

```python
# Drop original gender,category and state column 
df = df.drop(columns=['gender','category','state','merchant','first','last','street','city','job','trans_num','new_trans_date_trans_time','new_dob'])

df.head()
```

```python
# Check the data types of each column
print(df.dtypes)
```

```python
# Split features and target variable
x = df.drop(columns=['is_fraud'])
y = df['is_fraud']
```

```python
#print x
x.head()
```

```python
#print y
y.head()
```

```python
y.tail()
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Scale the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#in this I am using random forest classifier model
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest model
model = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
```

```python
# Train the model
model.fit(x_train_scaled, y_train)

# Predict on the test set
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}')

#Precision
precision = precision_score(y_test, y_pred, zero_division=1)
print(f'Precision: {precision*100:.2f}')

#recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall*100:.2f}')

#f1 score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1*100:.2f}')

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
#as recall is not good 
#also confusion matrix shows that many of frauds are detected non-fraud
#therefore we are resampling our dataset using SMOTE
from imblearn.over_sampling import SMOTE

# Use SMOTE to oversample the minority class (fraudulent transactions)
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train_scaled, y_train)

# Train the model again with the resampled dataset
model.fit(x_train_resampled, y_train_resampled)
```

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")

# Predict on the test set
y_pred = model.predict(x_test_scaled)
```

```python
#calculating after again resampling data
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}')

#Precision
precision = precision_score(y_test, y_pred, zero_division=1)
print(f'Precision: {precision*100:.2f}')

#recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall*100:.2f}')

#f1 score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1*100:.2f}')

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
#as recall is not good 
#also confusion matrix shows that many of frauds are detected non-fraud
#therefore we are resampling our dataset using RANDOM UNDERSAMPLING
from imblearn.under_sampling import RandomUnderSampler

# Use SMOTE to oversample the minority class (fraudulent transactions)
undersampler = RandomUnderSampler(random_state=42)
x_train_resampled, y_train_resampled = undersampler.fit_resample(x_train, y_train)

# Train the model again with the resampled dataset
model.fit(x_train_resampled, y_train_resampled)
```

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")

# Predict on the test set
y_pred = model.predict(x_test_scaled)
```

```python
#calculating for resampled data

#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}')

#Precision
precision = precision_score(y_test, y_pred, zero_division=1)
print(f'Precision: {precision*100:.2f}')

#recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall*100:.2f}')

#f1 score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1*100:.2f}')

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Train the model again with the resampled dataset
model.fit(x_train_resampled, y_train_resampled)

#using threshold tuning
y_pred_prob = model.predict_proba(x_test_scaled)[:, 1]
threshold = 0.45
y_pred = (y_pred_prob >= threshold).astype(int)
```

```python
#calculating after threshold tuning

#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}')

#Precision
precision = precision_score(y_test, y_pred, zero_division=1)
print(f'Precision: {precision*100:.2f}')

#recall
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall*100:.2f}')

#f1 score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1*100:.2f}')

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```
